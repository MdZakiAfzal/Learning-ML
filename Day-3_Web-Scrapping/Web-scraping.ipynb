```python
books = soup.find_all('li',class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
```

```python
name=[]
price=[]
available=[]
for i in books:
    name.append(((i.find('h3')).find('a'))['title'])
    price.append((i.find('p', class_='price_color').text)[1:])
    available.append((i.find('p', class_='availability').text).strip())

d = {'name':name, 'price':price, 'availability':available}
data = pd.DataFrame(d)
data
```
